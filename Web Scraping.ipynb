### Web scraping using pandas read html

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
```

```python
import pandas as pd
```

```python
html= pd.read_html(url, header = 0)
df2022=html[0]
df2022
```

```python
len(df2022)
```

```python
df2022[df2022.Age == 'Age'] ## identiying the rows which are multiple headers in the table
```

```python
len(df2022[df2022.Age == 'Age'])
```

##### Basic Data cleaning

```python
df = df2022.drop(df2022[df2022.Age == 'Age'].index) ## dropping all those multiple headers  present in the table
df
```

###### Quick Exploratory Analysis

```python
import seaborn as sns
```

###### Histogram showing the distribution of  Points per Game (PTS) 

```python
sns.distplot(df.PTS, kde=False, hist_kws=dict(edgecolor="black", linewidth = 1))
```

###### There are more than 100 players with 5 points per game. Number of players having 30 points per game are less than 10
